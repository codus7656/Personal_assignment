# 1. 소셜 네트워크 분석의 개요

- 네트워크는 각각의 개체들이 상호 **연결**된 구조를 말함. 일상생활에서 예를 들자면, **세계항공기 노선도**, **수도권 지하철 노선도** 그리고 **가계도**나 **메신저 대화**에서도 네트워크를 발견할 수 있음.
- 네트워크의 구성요소는 **node**와 **llink** 임.
- **노드는 고유한 속성을 가지고 있는 행위자**고 **링크는 노드들간의 관계**로, 쉽게 **지하철**을 예로 들자면, **각각의 전철역은 노드이고 전철간의 연결을 링크**라고 말할 수 있음.
- 이처럼 네트워크란 복수의 노드들이 연결된 구조임.




##  네트워크 이론의 개요 및 역사
- 그래프 이론이란 점과 그 점을 잇는 선으로 이루어진 그래프를 수학적으로 연구하는 이론
- 네트워크 이론은 수학의 그래프 이론에서 비롯된 이론

### 1. 오일러의 '쾨니히스베르크 다리 건너기 문제'
- 그래프 이론의 발전에 있어 탄탄한 수학적 기초를 제공
- 그래프 이론에서 **오일러 경로(Euler path, Eulerian path)** 는 **그래프의 모든 변을 단 한 번씩만 통과하는 경로**로 오일러 회로가 아닌 시작과 끝 꼭짓점이 다른 경로, 오일러 경로가 있을 필요충분조건은 **'정확히 두 개의 꼭짓점만이 홀수의 차수를 가지고 그래프가 연결되어 있다.'** 는 것.
- **오일러 회로(Euler circuit, Eulerian circuit)** 는 **그 가운데 같은 꼭짓점에서 시작해서 끝나는 오일러 경로**, 그래프가 오일러 회로를 가질 필요충분조건은 **'그래프가 연결된 그래프이고, 모든 꼭짓점의 차수가 짝수이어야 한다.'** 임
- **오일러 그래프** : 오일러 회로를 지닌 무향그래프
- **차수** : 각 꼭짓점에 연결된 변의 개수

### 그라노베터의 '약한 연결의 강점'
- 논문 **'약한 연결의 강점'** 에서 노동자들이 자신의 사회적 네트워크를 활용하여 직업을 어떻게 얻는지에 대해 조사하여 발표
- 직업을 얻게 된 대부분의 사람들은 가까운 사람들을 통해서가 아닌, 그냥 가볍게 알고 지내는 지인을 통해 얻은 경우가 훨씬 더 많았음
- **약한 사회적 관계의 유용성**에 대한 새로운 시사점을 제시

### 프리만의 '중심성 지표' 개발
- 각 노드의 중심성을 측정하는 세 가지 '중심성 지표' : **연결정도 중심성(Degree centrality), 근접 중심성(Closeness centrality), 매개 중심성(Betweenness centrality)**
-  매개 중심성은 한 노드가 다른 노드들간의 네트워크 관계에 있어서 중개자가 혹은 매개자 역할을 얼마나 수행하는지 측정하는 지

### 버트의 '구조적 공백'
- 구조적 공백(Structural holes)’ 이론은 사회 네트워크 구조 내에서 **상대적으로 중요한 위치와 역할을 하는 노드들**을 설명
- 구조적 공백의 위치에 있는 노드는 다른 노드들과 중복적인 관계를 가지고 있지 않기 때문에 **노드와 노드를 연결시켜 주는 역할**을 함
- 버트의 구조적 공백은 그라노베티의 ‘약한 연결(weak ties)’ 개념을 더 구체화 한 것으로 볼 수 있음.
- 실제 조직 내에서 형성된 개인의 인적 네트워크 구조상에서 구조적 공백이 많은 사람이 구조적 공백이 적은 사람의 경우보다 더 빨리 승진을 한다는 것을 밝혔음.
- 이와 같은 연구 결과는 구조적 공백이 **네트워크 내에서의 정보 흐름과 영향력에 있어 상대적으로 중요한 역할을 한다는 것**을 증명

### 왓츠, 스트로가츠의 '좁은 세상 네트워크'
- 인접한 노드와 일정한 규칙에 따라 일정한 숫자로만 연결되어 있는 정규 네트워크(Regular network), 무작위로 연결되어 있는 무작위 네트워크(Random network)의 중간에 존재하는 좁은 세상 네트워크(small world network)를 제시
- 노드들 간에 지름길이 존재. 이를 통해 전체 네트워크가 긴밀하게 연결.
- 이후 무척도 네트워크(scale-free network)가 밝혀지게 되는 계기를 제공

### 바라바시, 알버트, 정하웅의 '무척도 네트워크'
- 하이퍼 링크로 연결된 인터넷의 물리적 연결 구조에 대한 논문에서 인터넷 연결 수 분포가 정규분포를 따르지 않고 **멱함수법칙(Power law)** 을따르는 것을 밝혔음. 이러한 네트워크 구조를 **무척도 네트워크(Scale-free network)** 라고 함
- **무작위 네트워크**는 링크의 연결패턴이 평균값을 중심으로 **정규분포**를 이루는 반면, **무척도 네트워크**에서는 **소수의 링크만을 가지고 있는 수많은 노드들과 수많은 링크를 가지고 있는 소수의 노드들로 양분되는 것.**
- 이러한 멱함수 분포는 네트워크 형성에 있어서 **선호적 연결**이라는 독특한 현상에 기인하고 있음. 노드 간에 연결이 이루어 질 때 특정 노드에 대한 선호적 연결 현상이 발생. 이러한 현상이 반복되게 되면 **아주 많은 링크를 가진 소수의 노드들이 형성**되게 되고, 이러한 소수의 노드들은 전체 네트워크에서 허브 등의 중요한 역할을 하게 됨.



# 2. 네트워크 이론의 기초

## 2.1. 그래프의 기본 개념
- 네트워크를 표현하는 네트워크 모델링 과정에서 중요하게 사용되는 것이 그래프, 행렬
- 그래프 이론(Graph theory) : 객체들 간의 네트워크 관계 모델을 수학적으로 표현된 그래프로 나타내는 연구 분야
- 그래프는 **노드(Node, vertex)** 로 구성, 이러한 노드들은    **링크(Link, edge)** 로 연결되어 있음.


- **무방향(Undirected)** 네트워크: 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크 등
- **방향(Directed) 네트워크**: 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔로잉 네트워크 등

***무방향과 방향성이 있는 네트워크 그래프 그려보기***


```python
# 무방향 네트워크 그래프 그리기
import networkx as nx
#객체 생성
g1 = nx.Graph()
#노드 추가
g1.add_node('a')
g1.add_node('b')
g1.add_node('c')
g1.add_node('d')
#엣지 추가
g1.add_edge('a', 'd')
g1.add_edge('a', 'b')
g1.add_edge('a', 'c')
#그래프 그리기
nx.draw(g1, with_labels = True)
```

```python
# 방향 네트워크 그래프 그리기
#객체 생성
g2 = nx.DiGraph()

g2.add_edges_from([('a', 'b'), ('b', 'c'), ('d', 'c'), ('a', 'b')])

nx.draw(g2, with_labels = True)
```

- 네트워크 그래프에서 **노드 간 관계의 유무**만이 아니라 **관계의 정도(강도)** 를 함께 표시할 수 있음
- 이진 네트워크(binary network) 그래프 : 관계의 유무만을 표현한 경우
- 가중 네트워크(valued network) 그래프 : 관계의 정도를 함께 표시한 경우
- 관계의 정도를 나타내는 방법은 링크 옆에 **가중치를 직접 표시**하거나 가중치의 정도에 따라 **링크 선의 굵기를 다르게 표시**


***가중 네트워크 그래프 그리기***

```python
# 가중치 직접 (숫자로) 표현하기
g3 = nx.DiGraph() 

# a노드에서 b노드(가중치는 3), b노드에서 c노드 (가중치는 4), a노드에서 d노드 (가중치는 2)
g3.add_weighted_edges_from([('a', 'b', 3), ('b', 'c', 4),('a','d',2)])
#  a에서 b를 향하는 엣지 추가. 가중치는 6
g3.add_edge('a', 'c', weight = 6)

pos = nx.spring_layout(g3)
nx.draw(g3, pos = pos, with_labels = True)

# 가중치 그래프에 보이게 그리기
labels = nx.get_edge_attributes(g3,'weight')
nx.draw_networkx_edge_labels(g3, pos, edge_labels = labels);
```

## 2.2. 그래프의 표현 방법
- **인접행렬(Adjacency matrix) 방식, 노드 리스트(Node list)방식, 엣지 리스트(Edge list)방식**

- **인접행렬(Adjacency matrix)**: 인접한 노드들 간의 연결 관계를 행렬(Matrix)로 표현, 네트워크 상에서 노드의 개수가 n인 경우 인접 행렬은 nxn의 정방 행렬로 표현
- 행과 열의 이름은 노드를 나타내며, 행과 열이 교차하는 지점의 행렬 값은 관계(link)의 유무를 1과 0으로 나타냄. 
- 행의 이름은 링크가 출발하는 노드를 나타내고 열의 이름은 링크가 도착하는 노드를 나타냄. 
- 무방향 네트워크의 경우 모든 관계가 쌍방향이어서 방향성이 상쇄되는 경우와 동일하게 됨. 즉 인접행렬은 대칭적인 특징을 가짐

- **노드 리스트(Node list)**: 인접한 링크가 있는 노드들만을 표기하는 방법으로 노드들만의 목록
- 맨 첫번째 열은 그래프에 존재하는 노드들의목록을 나타내고 그 우측열 부터는 해당되는그 노드에 연결된 노드의 목록을 나타냄. 전체 행의 개수는 전체 노드의 개수와 일치하며, 노드별로 연결된 링크의 개수가 많을수록 열의 개수도 증가하게 됨. 방향성의 표시를 위해 incoming, outgoing을 구분함.

- **엣지 리스트(Edge list)**: 노드들을 연결한 링크들의 리스트로 “링크의목록＂
- 각각의 행은 링크를 통해서 연결되는 두 개의 노드 쌍. 따라서 리스트는 2열로만 이루어지고, 행의 개수는 전체 링크의 개수와 일치.
- 엣지 리스트에서는 그래프의 방향성을 자체적 으로 구분하기 어렵기 때문에 그래프의 방향성 여부를 별도로 명시해 주어야 함.
- 고립된 노드의 경우 링크가 존재하지 않으므로 목록에서 제외됨.

# 3. 네트워크 유형

### 3.1. 방향성과 가중치에 따른 분류
- 네트워크는 **링크의 방향성 유무**와 **가중치 유무**에 따라 크게 4가지 유형으로 구분
- (1)가중치 있음, 방향성 있음. (2) 가중치 있음, 방향성 없음. (3) 가중치 없음, 방향성 있음. (4) 가중치 없음, 방향성 없음.

### 3.2. 관계의 위상에 따른 분류
- 네트워크를 이루고 있는 **관계 위상 형태**에 따라서 구분
- ex) Ring, Mesh, Star, Fully connected, Line, Tree, Bus
- ex) Hub-and-Spoke, 여러 항공 노선이 대도시 터미널 집중 방식으로 운영

### 3.3. 네트워크 크기에 따른 분류
- 네트워크를 이루고 있는 **노드의 링크의 규모** 등 **네트워크 크기**에 따라 구분
- **노드 규모(수)**와 **링크 규모(수)** 에 따라 Micro(소규모), Meso(중규모), Macro(대규모), Super-macro(초대규모)로 분류할 수 있음.

### 3.4. 분석 대상에 따른 분류
- **전체 네트워크(complete network)**: 네트워크를 구성하고 있는 모든 노드를 포함한 네트워크를 말함. 완전 네트워크(whole network)라고도 함.
- **하위 네트워크(sub network)**: 전체 네트워크의 일부분으로 부속되어 있는 것으로 **컴포넌트(component), 파당(Clique), 클러스터(Cluster)** 등이 해당

- **파당(Clique)**: 네트워크의 하위 네트워크, 파당에 포함된 **모든 노드들이 상호간에 직접적으로 연결(완전연결)** 되어 있음. **다른 파당에는 포함되어 있지 않은 경우를 말함**. 일반적으로 양자 관계는 제외, **3개 이상의 노드를 포함한 것**을 파당이라고 함.

- **클러스터(Cluster)**: **유사한 속성(특성)을 가지고 있는 노드들의 집합**. 유사성에 따라 클러스터를 분류해 내는 작업을 클러스터링(Clustering)이라고 함. 노드간 유사성 분석을 통한 클러스터링은 수학적 알고리즘을 기반으로 계산 및 시각화되며 대부분의 소셜네트워크 알고리즘에 내장되어 있어 손쉽게 결과물을 얻을 수 있음. Clauset-Newman-Moore 알고리즘과 Girvan-Newman 알고리즘이 있는데 전자가 분석시간이 매우빠르고, 대규모 네트워크 분석이 가능. 하위 네트워크의 질적인 속성까지 고려하여 클러스터를 탐색함.
- ex) 미국 상원의원들이 의회의 표결 안건들에 대해 서로 동일한 표결을 한 정도를 네트워크로 나타냄.

- **에고 네트워크(Ego network)**: 네트워크 상에서 분석하고자 하는 **특정 노드를 중심에 두고, 그 노드와 직접적으로 연결된 노드들 간의 관계**를 네트워크로 표현한 것. 자아중심 네트워크(Ego-centric network)라고도 함.


# 4. 소셜 네트워크 분석 지표 
## 4.1. 네트워크 기본 속성 분석

### 네트워크 수준의 속성 분석

#### 1. 네트워크 크기(Network size)
- **네트워크를 구성하는 노드들의 수**
- 네트워크 크기가 증가할수록 네트워크 구조의 복잡성도 증가
- 연결관계의 수: 

방향성이 있는 경우: nx(n-1),
방향성이 없는 경우: nx(n-1)/2

- 소셜 네트워크 분석에 있어 중요한 역할

#### 2. 네트워크 밀도(Network density)
- **네트워크를 구성하고 있는 노드들 간의 연결된 정도
- 네트워크 내의 노드들 간의 연결이 많아질수록 밀도는 높아짐.
- 네트워크 밀도 계산:

무방향 이진 네트워크의 밀도: k/{n x (n-1)/2},
방향 이진 네트워크의 밀도: k/{n x (n-1)},
k: 실제 연결된 링크의 수, n: 노드의 수


#### 3. 포괄성(Inclusiveness)
- 전체 네트워크의 밀집도를 연결된 노드의 관점에서 표현
- **네트워크상의 총 노드의 개수 중에서 실제로 연결 관계가 존재하는 노드의 개수가 차지하는 비율
- 포괄성 계산: n-nisolated / n
- n: 총 노드의 수, nisolated: 고립된 노드의 수
- **동일한 네트워크 밀도를 가지고 있어도 포괄성은 차이가 있을 수 있음.

### 노드 수준의 속성분석

#### 4. 연결정도(Degree)
- 해당 노드에 직접 연결되어 있는 **노드들의 개수**, 또는 **링크의 개수**
- n개의 노드가 존재한다고 가정 했을 때, 각각의 노드에 최대 n-1개까지의 연결 정도가 가능
- 고립된 노드의 경우에는 연결정도가 0
- 연결정도의 측정을 통해 전체 네트워크 구조상에서 해당 노드가 보여주는 활동성과 영향력 유추가 가능
- 연결정도가 높을수록 다른 노드들과 많은 관계를 가지고 있음을 보여주며 이는 상대적으로 높은 활동성과 영향력을 반증함

#### 5. 연결강도(Strength)
- 노드간 관계(링크)의 강도
- **이진 네트워크(Binary network)**: 관계의 유무만 표시한 네트워크
- **가중 네트워크(Valued network)**: **연결 관계의 강도**를 상대적인 가중치를 함께 표시한네트워크. 관계 빈도, 친밀성, 지속성, 신뢰성 등 여러가지 관계 특성을 정의 가능

#### 6. 연결거리(Distance)
- 두개의 특정 노드들 사이에 존재하는 다양한 연결경로 중에서 가장 적은 수의 링크 단계를 거치는 경로, 즉 가장 짧은 연결 경로
- **최단 경로거리(Shortest path distance)**
- 두 노드간 연결 거리가 짧을수록 노드간의 연결이 높아짐
- 정보나 구전 등이 **얼마나 신속하게 전달**되는지 가늠하는 중요한 지표

#### 7. 직경(Diameter)
- 네트워크 상의 모든 연결거리 중에서 가장 긴 연결거리
- **네트워크 상에서 가장 멀리 떨어져있는 두개 노드의 연결거리=네트워크 지름**
- 네트워크 직경이 작을수록 신속한 정보 전달이 가능해지고 효율성 높아짐

#### 8. 평균연결거리(Average Distance)
- 네트워크내 모든 노드 쌍의 최단 연결거리를 산출하여 평균값을 계산한 것을 해당 네트워크의 평균 연결거리라고 함.
- 이것은 **네트워크 임의의 두 노드가 평균적으로 최소 몇 단계의 링크를 걸쳐서 연결이 가능한지** 보여줌

### 네트워크에 내재된 속성 분석

#### 9. 상호성(Reciprocity)
- 방향 네트워크 상에서 2개의 노드 간의 연결이 서로 **쌍방향**인 경우: **상호적 연결**, **단방향**인 경우 **비상호적연결**이라고 함.
- 상호성 계산은 전체 네트워크 상에서 상호적 연결이 차지하고 있는 비율로 산출


***상호적 연결일 경우,***

***노드 쌍 기준 상호성 : 상호적으로 연결된 노드 쌍/연결된 전체 노드 쌍***

***연결 관계(링크)기준 상호성: 상호적으로 연결된 링크 수/연결된 전체 링크 수***

- 아래 예시를 통해 계산해 보면,

```python
# 네트워크 그래프를 통해 상호적 연결이 차지하고 있는 비율 게산해보기
g2 = nx.DiGraph()

g2.add_edges_from([('a', 'b'), ('b', 'c'), ('c','b'), ('d', 'c'), ('a', 'b'),('c','d'),('a','e'), ('f', 'd'),('d','e'),('a','d')])

nx.draw(g2, with_labels = True)
```

- **노드 쌍 기준 상호성**을 게산해보면, 상호적으로 연결된 노드 쌍은 2개이고 연결된 전체 노드 쌍은 6개 이므로 0.33 이다. 따라서 **네트워크 상에 존재하는 모든 노드 중 33%가 상호적연결되어 있다는 뜻**이다.
- **연결 관계 기준 상호성**을 계산해보면, 상호적으로 연결된 링크 수는 4개이고 연결된 전체 링크 수는 8개 이므로 0.5이다. 따라서 **네트워크를 구성하고 있는 모든 링크 연결의 50%가 상호적 관계로 이루어져 있다는 것을 의미**한다.

#### 10. 군집화 계수(Clustering coefficient)
- **군집화(Clustering)**: 네트워크 상에서 특정 부분의 노드들이 밀접하게 연결되어 다른 부분보다 네트워크 밀도가 높아지는 현상
- **군집화 계수(Clustering coefficient)**: 군집화의 정도를 나타내는 지표를 뜻하며 해당 노드의 이웃 노드들 간에 가능한 최대 관계의 수 중에서 실제로 존재하는 관계의 수가 차지하는 비율로 정의
- 지역 군집화 계수

## 4.2. 중심성 분석
- **중심성 지표**: 소셜 네트워크 분석 지표 중에서 일반적으로 가장 많이 사용되는 지표로 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표
- 중심성 분석을 통해 전체 네트워크에서 **중요한 역할을 하는 노드**가 무엇인지 파악 가능
- 네트워크에서 중심적인 역할을 하는 구성원을 탐색하고 그 구성원을 통해 효율적인 정보 전달과 의사전달을 가능케함으로써 문제를 효과적으로 해결가능

#### 1. 연결정도 중심성(Degree centrality)
- 네트워크 상에서 **한 노드가 얼마나 많은 연결 관계를 가지고 있는 지를 측정하는 지표**
- 노드는 다른 노드들과 많은 연결 관계를 가질수록 더 넓은 선택의 폭과 자율성을 가지게 되어 다른 노드에 대한 의존성을 상대적으로 낮춤
- 결과적으로 네트워크 상에서 더 큰 영향력
- 노드에 연결된 링크의 수가 많아질수록 해당 노드의 연결 정도 중심성은 높아짐
- 연결정도 중심성은 절대적 연결정도 중심성과 상대적 연결정도 중심성을 나눌 수 있음 


```python
# 네트워크 그래프를 통해 연결정도 중심성 (Degree centrality) 알아보기
gf = nx.DiGraph()
gf.add_edges_from([('a', 'b'), ('b', 'd'), ('a', 'c'), ('a', 'd'), ('c', 'd'), ('d', 'f'),('g', 'd')])
nx.draw(gf, with_labels = True)
```

```python
# 한 노드가 얼마나 많은 연결 관계를 가지고 있는 지를 측정하는 지표
# 화살표를 받는 노드는 d가 가장 많은 연결 관계를 가지므로 중심성의 값이 0.8로 가장 높은 것을 확인할 수 있다.
degree_in=nx.in_degree_centrality(gf)
degree_in
```

```python
# 화살표를 주는 노드는 a가 가장 많은 연결 관계를 가지므로 중심성의 값이 0.6000000000000001로 가장 높은 것을 확인할 수 있다.
degree_out=nx.out_degree_centrality(gf)
degree_out
```

#### 2. 근접 중심성(Closeness centrality)
- 네트워크 상의 **노드들 간의 근접도를 기준으로 정의**
- 해당 노드가 전체 네트워크 상에서 얼마나 중앙에 위치하고 있는지를 측정
- 노드가 중앙에 가깝게 위치할 수록 자원과 정보를 신속하게 전체 네트워크로 확산
- 근접 중심성 계산 

#### 3. 매개 중심성(Betweenness centrality)
- 네트워크 상에서 **특정 노드가 다른 노드들의 중간에서 얼마나 중개자 및 매개자 역할을 하고 있는 지를 측정하는 지표**
- 노드의 매개 중심성이 높을 수록 네트워크 내에서 정보 및 자원의 흐름에 있어서 높은 통제력 확보
- 매개 중심성이 높은 노드는 관계 연결의 길목에서 정보를 왜곡하거나 관계 활동의 방해를 통해 네트워크를 위협할 수 있음
- 노드 i의 매개 중심성은 네트워크 내의 모든 노드 쌍의 최단 경로 중에서 노드 i가 포함되어 있는 경로의 비율로 정의되며, 이것을 절대적 매개중심성이라고 함. 상대적 매개중심성은 이렇게 계산한 절대적 매개 중심성 값을 이론상 가능한 매개 중심성 최댓값으로 나누어 계산함. 


#### 4. 아이겐 벡터 중심성(Eigenvector centrality)
- 해당 노드에 직접 연결된 다른 노드들의 개수 뿐만 아니라 **연결된 다른 노드들의 중요도 또한 함께 반영한 것**
- 중심성이 높은 노드들과 많이 연결된 노드일수록 해당 노드의 아이겐벡터 중심성은 커짐
- 네트워크 상에서 소수 몇몇 영향력 높은 노드들에 연결되어 있는 것이 다수의 일반 노드들에 연결되어 있는 것보다 더 큰 영향력을 가짐
- 즉 친구가 많은 것도 중요하지만 **중요한 친구가 많은 지**를 보고자 할 때 사용하는 지표임

## 피처링에 대한 정보로 네트워크 그래프 그리기

```python
# 필요한 패키지 불러오기
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# csv 파일 불러와 읽기
# Bugs 사이트에서 50곡의 가수들의 피쳐링에 대한 정보를 저장함.
# from열이 곡의 가수, to열이 피쳐링을 해준 가수
csv1 = pd.read_csv('featuring.csv', encoding='utf-8')
csv1

# csv 파일에서 from열과 to열을 각각 start, end라는 변수에 tuple형태로 저장하기
start = tuple(csv1['from'])
end = tuple(csv1['to'])


# 빈 리스트 만들기
list=[]

# start와 end변수에 저장되어 있는 깂을 차례로 저장하여 튜플 값을 갖는 리스트 만들기
for i in range(0,50): 
    a=(start[i],end[i])
    list.append(a)

# featuring 네트워크 그래프 생성하기
featuring = nx.DiGraph()
featuring.add_edges_from(list)

# 영향력에 따라 노드 크기를 다르게 하여 그래프 생성하기
d = dict(featuring.degree)
nx.draw(featuring, node_size = [v * 100 for v in d.values()], with_labels = True)


```

```python
# 연결정도 중심성 확인해보기
de_in=nx.in_degree_centrality(featuring)
de_in

# 화살표를 받는 노드는 'The Quiett','Crush','DEAN'이  가장 많은 화살표를 받음으로 0.08로 가장 크다.
```

```python
# 연결정도 중심성 확인해보기
de_out=nx.out_degree_centrality(featuring)
de_out

# 화살표를 주는 노드는 ''Epik High'이 가장 많은 화살표를 주므로 0.12로 가장 크다.
```
